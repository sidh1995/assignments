```python
# extracting the records where a team was batting first
batting_first = ipl[ipl['win_by_runs'] != 0]
batting_first
```

```python
batting_first.head() 
```

```python
# making a histogram
plt.figure(figsize = (5,5))
plt.hist(batting_first['win_by_runs'])
plt.title('distribution of runs')
plt.xlabel('runs')
plt.show()
```

```python
# finding out the nuber of wins each team after  batting first
batting_first['winner'].value_counts()
```

```python
# making a bar-plot for top 5 teams with most wins after batting first

plt.figure(figsize = (14,15))
plt.bar(list(batting_first['winner'].value_counts()[0:5].keys()),list(batting_first['winner'].value_counts()[0:5]),color = ['blue','yellow','orange','pink','green'])
plt.show()
```

```python
# making pie chart

plt.figure(figsize = (7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
# extracting those records where a team has won after batting second 
batting_second =  ipl[ipl['win_by_wickets'] != 0]
```

```python
batting_second.head()
```

```python
# making a histogram for frequency of wins W.r.t number of wickets
plt.figure(figsize=(5,5))
plt.hist(batting_second['win_by_wickets'],bins = 30)
plt.show()
```

```python
# finding out the frequency of nuber of wins W.r.t each time after batting second
batting_second['winner'].value_counts()
```

```python
# making a bar for top-3 teams with most wins after batting second
plt.figure(figsize = (7,5))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color =['blue','red','green'])
plt.show()
```

```python
#  making a pie chart for distribution of most wins after batting second

plt.figure(figsize=(5,5))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
